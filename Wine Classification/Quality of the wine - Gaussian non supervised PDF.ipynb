```python
from sklearn import tree
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import interactive
interactive(True)
import seaborn as sns
from scipy.stats import multivariate_normal

from IPython.display import SVG
from graphviz import Source
from IPython.display import SVG, display
from IPython.display import HTML
```

```python
datafRed = pd.read_csv("winequality-red.csv", sep=";")
datafRed.head()
```

```python
redMean = datafRed.mean()
```

```python
redCov = datafRed.cov()
redCov
```

```python
for i, row in datafRed.iterrows():
    guess = multivariate_normal.pdf(row,mean=redMean, cov=redCov)
    if guess<0.05:
        print(guess, "\nExceptional(the worst or the best):\n",row)
```

```python
datafWhite = pd.read_csv("winequality-white.csv", sep=";")
datafWhite.head()
```

```python
whiteMean = datafWhite.mean()
whiteCov= datafWhite.cov()
```

```python
for i, row in datafWhite.iterrows():
    guess = multivariate_normal.pdf(row,mean=whiteMean, cov=whiteCov,allow_singular=True)
    if guess<0.05:
        print(guess, "\nExceptional(the worst or the best):\n",row)
```
